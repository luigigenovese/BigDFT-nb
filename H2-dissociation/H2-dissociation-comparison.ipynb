# Comparison between the different approximation for H2 dissociation

```python
from BigDFT import Logfiles as lf

HtoeV = 27.211396132 #Conversion Hartree to meV

distances = []
crmult = []

multi = dict()
ref_multi = dict()

def collect_data(key):
    ref_multi[key] = {'rmult': [], 'energies': [], 'gnrm_cv': [], 'gnrm': [], 'legend': 'H-%s' % key}
    multi[key] = {'rmult': [], 'distances': [], 'energies': [], 'forcemax': [], 'gnrm': [], 'legend': 'H2-%s' % key}
    ref = ref_multi[key]
    data = multi[key]
    arch = lf.Logfile(archive='log-H2-%s-dissociation.tar.bz2' % key)
    for log in arch:
        if log.nat == 1:
            r = float(log.log['dft']['rmult'][0])
            ref['rmult'].append(r)
            ref['gnrm_cv'].append(log.gnrm_cv)
            ref['gnrm'].append(lf.find_iterations(log.log)[-1][-1])
            ref['energies'].append( HtoeV*log.energy)
        else:
            r = float(log.log['dft']['rmult'][0])
            d = float(log.log['posinp']['positions'][1]['H'][2])
            if r not in crmult: 
                crmult.append(r)
            if d not in distances:
                distances.append(d)
            data['rmult'].append(r)
            data['distances'].append(d)
            data['energies'].append( HtoeV*log.energy )
            data['gnrm'].append(lf.find_iterations(log.log)[-1][-1])
            data['forcemax'].append(log.forcemax)
            
#Collect the data for the LDA
collect_data('LDA')
```

```python
#Collect the data of Hartree-Fock approximation.
collect_data('HF')
```

```python
#Collect the data of local spin density approximation (LSDA).
collect_data('LSDA')
```

```python
#Collect the data of LSDA with input guess up in one atom and down in the other one
collect_data('LSDA-IGSpin')
```

## Total energy for the Hydrogen atom in function of rmult

```python
import matplotlib.pyplot as plt
import numpy as np
# Lists of markers and colors (for matplotlib)
colors = ['#74a9cf', '#2b8cbe', '#045a8d', '#009900', '#FF8000']
colors = ['#000000', '#ff0000', '#045a8d', '#009900', '#FF8000']
markers = ['o','s','d','d','d']

plt.figure(figsize=(15,7))
plt.xlabel('Extension of the grid')
plt.ylabel('Total Energy (Hartree)')
for i,ref in enumerate(ref_multi.values()):
    ener = ref['energies'][-1]
    print "The Total energy for '%s' is %0.4f eV (%0.6f Hartree)." % (ref['legend'], ener, ener/HtoeV)
    plt.plot(ref['rmult'], [r/HtoeV for r in ref['energies']], marker=markers[i], 
             ls='-', label=ref['legend'], color=colors[i])
plt.legend(loc=1)
plt.show()
```

# H2 dissociation energy curve for different functionals

```python
#Build the curve
curves = [ {'energies': [], 'distances': [], 'legend': "H2 %s" % m} for m in multi.keys()]
for i,(key,data) in enumerate(multi.items()):
    for (r,d,e) in zip(data['rmult'],data['distances'],data['energies']):
        if key == 'LSDA':
            eref = ref_multi['LDA']['energies'][-1]
        else:
            eref = ref_multi[key]['energies'][-1]
        if r == 12.0:
            curves[i]['energies'].append(e-2.0*eref)
            curves[i]['distances'].append(d)

import matplotlib.pyplot as plt
import numpy as np
# Lists of markers and colors (for matplotlib)
colors = ['#74a9cf', '#2b8cbe', '#045a8d', '#009900', '#FF8000']
colors = ['#000000', '#ff0000', '#045a8d', '#009900', '#FF8000']
markers = ['o','s','d','d','d']

plt.figure(figsize=(15,7))
# Plot with matplotlib
for i, dat in enumerate(curves):
    im = i%len(colors)
    if i == 0: plt.plot(dat['distances'],np.zeros_like(dat['energies']),'k-')
    plt.plot(dat['distances'], dat['energies'], marker=markers[im], 
             ls='-', label=dat['legend'],color=colors[im])  

plt.ylim([-8, 10])
#plt.xlim([6.0, 10.0])
plt.xlabel('Distance ($\AA$)')
plt.ylabel('E(method)-$2E_{ref}$(method) (eV)')
plt.title('Dissociation energy of the H2 dimer for different functionals')
plt.legend(loc=1)
plt.show()
```

# Curve of Gunnarsson and Lundqvist (JCP 1976)

The reference for two isolated hydrogen atom is -1 Hartree.

```python
#Build the curve
curves = [ {'energies': [], 'distances': [], 'legend': "H2 %s" % m} for m in multi.keys()]
for i,(key,data) in enumerate(multi.items()):
    rc = 12.0
    if key == 'LDA':
        eref = ref_multi['LSDA']['energies'][-1]
    elif key == 'HF':
        eref = ref_multi['HF']['energies'][-1]
    else:
        eref = ref_multi[key]['energies'][-1]
    for (r,d,e) in zip(data['rmult'],data['distances'],data['energies']):
        if r == rc:
            curves[i]['energies'].append(e-2.0*eref)
            curves[i]['distances'].append(d)

import matplotlib.pyplot as plt
import numpy as np
# Lists of markers and colors (for matplotlib)
colors = ['#74a9cf', '#2b8cbe', '#045a8d', '#009900', '#FF8000']
colors = ['#000000', '#ff0000', '#045a8d', '#009900', '#FF8000']
markers = ['o','s','d','d','d']

plt.figure(figsize=(15,7))
# Plot with matplotlib
for i, dat in enumerate(curves):
    im = i%len(colors)
    if i == 1: plt.plot(dat['distances'],np.zeros_like(dat['energies']),'k-')
    plt.plot(dat['distances'], dat['energies'], marker=markers[im], 
             ls='-', label=dat['legend'],color=colors[im])  

plt.ylim([-5.0, 3.0])
#plt.xlim([0.0, 2.0])
plt.xlabel('Distance ($\AA$)')
plt.ylabel('$E(method)-2E_{ref}$ (eV)')
plt.title('Dissociation energy of the H2 dimer for different functionals')
plt.legend(loc=4)
plt.show()
```
